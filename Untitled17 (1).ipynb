While training any model, the independent variables can not be string so converting them into categorical variable and using one hot encoding method to encode the string values to categorical values like 0,1,2...

```python
from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
df['Team_Cat'] = labelencoder.fit_transform(df['Team'])
df['Name_Cat'] = labelencoder.fit_transform(df['Name'])
df['Round_Cat'] = labelencoder.fit_transform(df['Round'])
df['Home_Team_Cat'] = labelencoder.fit_transform(df['Home Team'])
df['Away_Team_Cat'] = labelencoder.fit_transform(df['Away Team'])
```

Performing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
modelLR = LogisticRegression(C = 1, tol = 0.00001, random_state = 0,max_iter=1000)
modelLR.fit(x_train,y_train)
```

```python
y_pred_lr = modelLR.predict(x_test)
print('Percentage correct: ', 100*np.sum(y_pred_lr == y_test)/len(y_test))
```

```python
Votes=[0,1,2,3]
from sklearn import metrics
lv = [Votes[i] for i in y_test.astype('int')]
yp = [Votes[i] for i in y_pred_lr.astype('int')]
print("Classification report for classifier %s:\n%s\n" % (modelLR, metrics.classification_report(lv, yp)))
```

```python
df = pd.DataFrame(
    np.c_[y_test, y_pred_lr],
    columns=['true_label', 'prediction']
)
from sklearn.metrics import confusion_matrix
label_names = Votes
lv = [Votes[i] for i in y_test.astype('int')]
yp = [Votes[i] for i in y_pred_lr.astype('int')]
cmx = confusion_matrix(lv, yp)
df1 = pd.DataFrame(cmx, columns=label_names, index=label_names)
df1.columns.name = 'prediction'
df1.index.name = 'label'
df1
```

As logistic regression is good for binary responses, this modelling is not good in our  case as we have 4 responses and not a binary one. It is either predicting 0 or 3 ,i.e., taking either 0 or 3 as its response value whereas we have 1 and 2 as well. So logistic regression is not suitable in this case.
